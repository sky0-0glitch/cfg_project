```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
data=pd.read_csv('education_career_success.csv')
```

```python
data = data.sort_values(by="University_GPA", ascending=True)
min_GPA = min(data.University_GPA)
max_GPA = max(data.University_GPA)
min_salary = min(data.Starting_Salary)
max_salary = max(data.Starting_Salary)
```

```python
xpoints = np.array(data.University_GPA)
ypoints = np.array(data.Starting_Salary)
plt.scatter(xpoints, ypoints, alpha=0.2, s=10, color='blue', label="Data Points")
z = np.polyfit(xpoints, ypoints, 1)
p = np.poly1d(z)
plt.plot(xpoints, p(xpoints), color='red', linestyle="dashed", label="Trend Line")
plt.xlabel('GPA')
plt.legend()
plt.ylabel('Salary')
plt.title("Relation between GPA and salary")
plt.text(0.5,1, f"Minimum GPA is {min_GPA} and maximum GPA is {max_GPA} and Minimum salary is {min_salary} and maximum salary is {max_salary}",fontsize=10, bbox=dict(facecolor='white', alpha=0.7))
plt.show
```
